```python
import time

def checkTime(func):
    def new_func(*args):
        start = time.time();
        print("--------------------------------------------------------------------------")
        print(time.strftime("[%Y-%m-%d %H:%M]"))
        func(*args)
        print("실행시간은:",time.time()-start)
        print("--------------------------------------------------------------------------\n")
    return new_func

class Person(object):
    def __init__(self,name,age,gender,address):
        self._Name = name
        self._Age = age
        self._Gender = gender
        self._Address = address

    def aboutMe(self):
        print("이름: "+self._Name)
        print("나이: "+self._Age)
        print("성별: "+self._Gender)
        print("주소: "+self._Address)
        
class Employee(Person):
    employee_num = 0
    def __init__(self,name,age,gender,address,salary,hiredate):
        Person.__init__(self,name,age,gender,address)
        salary = int(salary)
        self._Salary = salary
        self._Salary_str = Employee.compute_Salary(salary)
        self._Hiredate = hiredate
        Employee.employee_num += 1
        self._ID = str(Employee.employee_num)
    
    def aboutMe(self):
        Person.aboutMe(self)
        print("ID: "+self._ID)
        print("연봉: "+self._Salary_str)
        print("입사일: "+self._Hiredate)
        
    
    @staticmethod
    def compute_Salary(salary):
        Salary_str = ""
        if(salary//100000000!=0):
            Salary_str += str(salary//100000000) + '억'
            salary = salary%100000000
        if(salary//10000!=0):
            Salary_str += str(salary//10000) + '만'
            salary = salary%10000
        if(salary!=0):
            Salary_str += str(salary)+'원'
        else:
            Salary_str +='원'
        return Salary_str
    
    @property
    def salary(self):
        return self._Salary
    
class Manager(Employee):
    manager_num = 0
    def __init__(self,name,age,gender,address,salary,hiredate,department,office):
        Employee.__init__(self,name,age,gender,address,salary,hiredate)
        self._Department = department
        self._Office = office
        Manager.manager_num += 1
    
    @checkTime
    def aboutMe(self):
        print("***Manager***")
        Employee.aboutMe(self)
        print("부서: "+self._Department)
        print("사무실: "+self._Office)
        print()
    
    @staticmethod
    def aboutNum():
        print("{}명의 직원들 중 {}명이 매니저 입니다.".format(Employee.employee_num,Manager.manager_num))

class Staff(Employee):
    def __init__(self,name,age,gender,address,salary,hiredate,department,manager):
        Employee.__init__(self,name,age,gender,address,salary,hiredate)
        self._Department = department
        self._Manager = manager
    
    @checkTime
    def aboutMe(self):
        print("***Staff***")
        Employee.aboutMe(self)
        print("부서: "+self._Department)
        print("매니저: "+self._Manager)
        print()
        
    
class Hourly(Employee):
    def __init__(self,name,age,gender,address,salary,hiredate,department,manager,hours_per_week):
        Employee.__init__(self,name,age,gender,address,salary,hiredate)
        self._Department = department
        self._Manager = manager
        self._Hour_per_week = hours_per_week
        weekly_wage = int(self.salary/12/4)
        self._Weekly_wage = Employee.compute_Salary(weekly_wage)
    
    @checkTime
    def aboutMe(self):
        print("***Hourly***")
        Employee.aboutMe(self)
        print("부서: "+self._Department)
        print("매니저: "+self._Manager)
        print("노동시간(주): "+self._Hour_per_week+"시간")
        print("주급: "+self._Weekly_wage)
        print()
        
```
