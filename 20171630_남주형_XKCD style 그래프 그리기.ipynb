```python
import matplotlib.pylab as plt
import numpy as np

X = np.linspace(-np.pi,np.pi,256)
C = np.cos(X)

with plt.xkcd():
    plt.title('XKCD style plot!!!')
    plt.plot(X,C)
    plt.scatter([2*np.pi/3],[np.cos(2*np.pi/3)],color="red")
    plt.annotate(r'0.5 Here!',xy = (2*np.pi/3,np.cos(2*np.pi/3)),xycoords = 'data',xytext = (-90,-50),textcoords='offset points',fontsize=16,arrowprops=dict(arrowstyle="->",linewidth=1,color="black"))
```

```python
import matplotlib.pylab as plt
import numpy as np

with plt.xkcd():
    fig = plt.figure()
    x1 = np.linspace(-10,10,256)
    y1 = x1-x1+10
    plt.plot(x1,y1)
    plt.ylim(-1,13)
    x2 = np.linspace(10,18,256)
    y2 = 23-1.3*x2
    plt.plot(x2,y2,color = "steelblue")
    plt.xticks([])
    plt.yticks([])
    plt.ylabel('my overall health')
    plt.xlabel('time')
    plt.annotate('THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',xy=(10,10),xycoords='data',xytext=(-200,-60),
                 textcoords='offset points', fontsize=16,arrowprops=dict(arrowstyle="->",linewidth=1,color="black"))
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    fig.text(0.5,0,"\"Stove Ownership\" from xkcd by Randall Monroe",ha='center')
```
