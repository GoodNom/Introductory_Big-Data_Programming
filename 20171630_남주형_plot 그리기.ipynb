```python
from pylab import *

X = linspace(-np.pi,np.pi,256)
C = cos(X)
plot(X,C)
xticks([-pi, -pi/2, 0, pi/2, pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1,0,1])
show()
```

```python
from pylab import *

X = linspace(-pi,pi,256)
C, S = cos(X), sin(X)
plot(X,C)
plot(X,S)
xticks([-pi, -pi/2, 0, pi/2, pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1,0,1])
show()
```
